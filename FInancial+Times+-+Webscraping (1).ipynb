##  Single Scrape 

```python
# pick up text from the first page of the Bitcoin search
response = requests.get('https://www.ft.com/search?q=bitcoin&page=1')

# Henter hjemmesiden ned i et format,der kan bruges 
soup = BeautifulSoup(response.text,"lxml" )

#Finds list to of articles.
page = soup.find_all('a', attrs={'class':'js-teaser-heading-link'})

page
```

```python
# Finds a list with links on ONE page
base_url= 'www.ft.com'
FT_links = []
for links in page:
    article_link = base_url + links['href']
    FT_links.append(article_link)
# Make all links searchable
linkslist = [w.replace('www.ft.comhttp://', 'http://') for w in FT_links]

linkslist
    
```

## Find all articlepages-links - full scrape 

```python
# This function make a list with all search result
def next_page():
    url = 'http://www.ft.com/search?q=bitcoin&page='
    FTpages= []
    for i in range(1,48):
        nextpage =  url + str(i)      
        FTpages.append(nextpage)
    return FTpages

FTpages = next_page()
print(FTpages)

# pick up text from the first page of the Bitcoin search
FT_links = []
for b in FTpages:
    response = requests.get(b)
    # Henter hjemmesiden ned i et format,der kan bruges 
    soup = BeautifulSoup(response.text,"lxml" )
    #Finds list to of articles.
    page = soup.find_all('a', attrs={'class':'js-teaser-heading-link'})
    base_url= 'www.ft.com'
    for links in page:
        article_link = base_url + links['href']
        FT_links.append(article_link)
        # Make all links searchable
    linkslist = [w.replace('www.ft.comhttp://', 'http://') for w in FT_links]
len(linkslist)

########tilføj http:// for alle linksene
  
```
