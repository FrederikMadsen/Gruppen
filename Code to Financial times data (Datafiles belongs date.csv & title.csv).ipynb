```python
#Loading dataset for dates
import pandas as pd
df = pd.read_csv('date.csv', sep='delimiter', dtype={'date':str}, engine='python')
df.head()

#Loading dataset for titles
df2 = pd.read_csv('title.csv', sep='delimiter', dtype={'title':str}, engine='python')
df2.head()

#Merges date & title dataset
df3 = df2.merge(df, left_index=True, right_index=True, how='outer')

#Change the date variable to dateindex (temporal data)
datetime_index = pd.to_datetime(df3.date)
df3.date = pd.Series(datetime_index)
```
